```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample attention scores (replace with your actual data)
# attention_scores = [0.2, 0.4, 0.6, 0.8, 0.9, 0.7, 0.5, 0.3]
attention_scores = [101, 1045, 4149, 1037, 2047, 18059, 1998, 2049, 18059, 1060, 2013, 6207, 102]

# Create a list of tokens corresponding to the attention scores
# tokens = ["Token 1", "Token 2", "Token 3", "Token 4", "Token 5", "Token 6", "Token 7", "Token 8"]
tokens = ["[CLS]", "I", "bought", "a", "new", "iphone", "and", "its", "Iphone", "X", "from", "Apple", "[SEP]"]

# Reshape the attention scores into a 2D matrix
num_tokens = len(tokens)
scores_matrix = np.array(attention_scores).reshape(1, num_tokens)

# Create a heatmap with annotations
plt.figure(figsize=(10, 2))
ax = sns.heatmap(scores_matrix, cmap='coolwarm', annot=True, fmt=".2f", xticklabels=tokens, yticklabels=False)

plt.show()
```

# Trying to get background colored text visualization 

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample attention scores (replace with your actual data)
# attention_scores = [0.2, 0.4, 0.6, 0.8, 0.9, 0.7, 0.5, 0.3]
attention_scores = [101, 1045, 4149, 1037, 2047, 18059, 1998, 2049, 18059, 1060, 2013, 6207, 102]

# Create a list of tokens corresponding to the attention scores
# tokens = ["Token 1", "Token 2", "Token 3", "Token 4", "Token 5", "Token 6", "Token 7", "Token 8"]
tokens = ["[CLS]", "I", "bought", "a", "new", "iphone", "and", "its", "Iphone", "X", "from", "Apple", "[SEP]"]

# Define a colormap for background colors (e.g., 'coolwarm' or 'viridis')
cmap = plt.get_cmap('coolwarm')

# Normalize attention scores to map them to the color scale
normalized_scores = (np.array(attention_scores) - np.min(attention_scores)) / (np.max(attention_scores) - np.min(attention_scores))

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 2))

# Iterate through tokens and attention scores, and plot colored text boxes
for i in range(len(tokens)):
    token = tokens[i]
    score = normalized_scores[i]
    color = cmap(score)  # Map the normalized score to a color
    ax.text(i, 0, token, ha='center', va='center', fontsize=12, bbox=dict(facecolor=color, edgecolor='k'))

# Set x-axis labels
ax.set_xticks(np.arange(len(tokens)))
ax.set_xticklabels(tokens, rotation=45)

# Remove y-axis
ax.get_yaxis().set_visible(False)

plt.show()
```
