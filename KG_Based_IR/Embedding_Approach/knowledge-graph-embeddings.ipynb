```python
import os
import pandas as pd
import numpy as np
from node2vec import Node2Vec
import networkx as nx

column_names = ['Filename', 'Embedding']
graph_embeddings = []

directory_path = '/kaggle/input/aila-docs-kg-dataset/KG_AILA_Docs/'
file_extension = '.csv'

document_embedding = []
document_ids = []

# files = ["KG_C9.csv", "KG_C10.csv", "KG_C11.csv"]

for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)

    if True:
        print(filename)
        doc_graph_df = pd.read_csv(file_path)
        if doc_graph_df.empty:
            print(f"Skipping {filename} since it has no rows.")
            continue
        G = nx.Graph()
        G.add_nodes_from(doc_graph_df["n"].unique())
        G.add_nodes_from(doc_graph_df["m"].unique())
        rel_edges = [(s, t, {"r_type": r}) for r, s, t in zip(doc_graph_df["r_type"], doc_graph_df["n"], doc_graph_df["m"])]
        G.add_weighted_edges_from(rel_edges)
        
        node2vec = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=4, weight_key=None)
        model = node2vec.fit(window=10, min_count=1, batch_words=4)
        embeddings = np.array([model.wv[node] for node in G.nodes()])
        document_embedding.append(embeddings)
        document_ids.append(filename)
        graph_embeddings.append((filename, embeddings))
#         graph_embeddings_df = pd.concat([graph_embeddings_df, pd.Series([filename, embeddings])], ignore_index=True)
```

```python
graph_embeddings = np.array(graph_embeddings, dtype=[('id', 'U10'), ('embedding', object)])
np.save('/kaggle/working/graph_embeddings.npy', graph_embeddings)
```
