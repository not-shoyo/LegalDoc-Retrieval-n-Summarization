```python
documents_per_query = {}
queries_per_document = {}

relevant_pairings = pd.DataFrame(query_document_relevance_pairings.loc[query_document_relevance_pairings['Relevance'] == 1])

for idx, pair in relevant_pairings.iterrows():
    queryName, documentName = pair["Query_Name"], pair["Document_Name"]
    if queryName not in documents_per_query:
        documents_per_query[queryName] = []
    if documentName not in queries_per_document:
        queries_per_document[documentName] = []
    documents_per_query[queryName].append(documentName)
    queries_per_document[documentName].append(queryName)

print(len(documents_per_query))
print(len(queries_per_document))
```

```python
# import statistics
from collections import Counter
import matplotlib.pyplot as plt

# Get the number of documents per query
num_documents_per_query = [len(docs) for docs in documents_per_query.values()]

# mode_value = statistics.mode(num_documents_per_query)

# Use Counter to find the mode and count occurrences
counter = Counter(num_documents_per_query)
mode_value = counter.most_common(1)[0][0]
num_queries_with_mode_value = counter[mode_value]

print(f"most common #documents for a query: {mode_value}")
print(f"number of queries with 'mode' documents: {num_queries_with_mode_value}")
print(f"total #queries: {len(documents_per_query)}")

# Plot the distribution
plt.bar(range(len(num_documents_per_query)), num_documents_per_query, align='center')
plt.xticks(range(len(num_documents_per_query)), documents_per_query.keys())
plt.xlabel('Queries')
plt.ylabel('Number of Documents')
plt.title('Distribution of Number of Documents per Query')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the number of documents per query
num_queries_per_document = [len(docs) for docs in queries_per_document.values()]

# Plot the distribution
plt.bar(range(len(num_queries_per_document)), num_queries_per_document, align='center')
plt.xticks(range(len(num_queries_per_document)), queries_per_document.keys())
plt.xlabel('Documents')
plt.ylabel('Number of Queries')
plt.title('Distribution of Number of Queries per Document')
plt.show()
```
