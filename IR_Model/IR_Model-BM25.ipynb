```python
document_names = documents_df["Document_Name"].str.strip('.txt')
document_names.head()
```

```python
relevant_query_document_pairings = pd.DataFrame(query_document_relevance_pairings.loc[query_document_relevance_pairings['Relevance'] == 1])
relevant_query_document_pairings
```

```python
count = 0
for i in range(50):
    query_name = queries_df.loc[i, "Query_Name"]
    relevant_document_names = set(relevant_query_document_pairings.loc[relevant_query_document_pairings['Query_Name'] == query_name]["Document_Name"])
    top_n = bm25.get_top_n(queries_df.loc[i, "preprocessed_text"].split(" "), document_names, n=10)
    count += len(relevant_document_names & set(top_n))

print(count)

Precision = count/500
Recall = count/195

print(Precision)
print(Recall)
```
