```python
import pandas as pd
from rank_bm25 import BM25Okapi

def bm25_search(query, queries_df, documents_df, top_n=10):
    # Tokenize the query
    tokenized_query = query.split()

    # Tokenize and preprocess the text in documents_df
    tokenized_documents = [doc.split() for doc in documents_df['preprocessed_text']]

    # Initialize BM25 object
    bm25 = BM25Okapi(tokenized_documents)

    # Get BM25 scores for the query
    scores = bm25.get_scores(tokenized_query)

    # Add scores to documents_df
    documents_df['bm25_score'] = scores

    # Sort documents based on BM25 scores
    ranked_documents = documents_df.sort_values(by='bm25_score', ascending=False)

    # Get the top N relevant documents
    top_documents = ranked_documents.head(top_n)

    return top_documents
```

```python
def perform_bm25(query):
    top_relevant_documents = bm25_search(query, queries_df, documents_df, top_n=20)
    return top_relevant_documents[['preprocessed_text', 'bm25_score']]

def ground_truth(query):
    relevant_document_names = relevant_query_document_pairings.loc[relevant_query_document_pairings['Current_Case_Name'] == "current_case_0001"]['Prior_Case_Name'].values
    relevant_doc_texts = []
    for document_name in relevant_document_names:
        relevant_doc_texts.append(documents_df.loc[documents_df['Prior_Case_Name'] == f"{document_name}.txt"][['preprocessed_text']].values)
    return relevant_doc_texts

def calc_total_right_returned(top_relevant_documents_texts, actual_relevant_documents):
    total_right_returned = 0
    for target_preprocessed_text in actual_relevant_documents:
        for returned_text in top_relevant_documents_texts:
            if returned_text == target_preprocessed_text:
                total_right_returned += 1
    return total_right_returned

def calculate_accuracy(top_relevant_documents_texts, actual_relevant_document_texts):
    total_returned = len(top_relevant_documents_texts)
    # print(total_returned)
    total_right_returned = calc_total_right_returned(top_relevant_documents_texts, actual_relevant_document_texts)
    # print(total_right_returned)
    return total_right_returned / total_returned

def calculate_precision(top_relevant_documents_texts, actual_relevant_document_texts):
    total_right = len(actual_relevant_document_texts)
    # print(total_right)
    total_right_returned = calc_total_right_returned(top_relevant_documents_texts, actual_relevant_document_texts)
    # print(total_right_returned)
    return total_right_returned / total_right
```

```python
import pandas as pd

def calculate_accuracy_precision(query, top_relevant_document_texts):
    # Assuming you have a function to determine relevant documents for a query
    relevant_documents_gt = ground_truth(query)  # Replace with your ground truth function
    
    # print(relevant_documents_gt)
    
    actual_relevant_document_texts = relevant_documents_gt

    # print(actual_relevant_document_texts)

    # Calculate accuracy
    accuracy = calculate_accuracy(top_relevant_document_texts, actual_relevant_document_texts)
    
    # Calculate precision
    precision = calculate_precision(top_relevant_document_texts, actual_relevant_document_texts)
    
    return accuracy, precision

def calculate_avg_accuracy_precision(queries_df, documents_df):
    avg_accuracy = 0.0
    avg_precision = 0.0
    individual_accuracies = []
    individual_precisions = []
    
    for index, query_row in queries_df.iterrows():
        query = query_row['preprocessed_text']
        
        # Get relevant documents using BM25
        top_relevant_documents = perform_bm25(query)
        
        # print(top_relevant_documents)
        # print(query)
        
        top_relevant_document_texts = top_relevant_documents['preprocessed_text'].values
        query_text = query

        # print(top_relevant_document_texts)
        # print(query_text)

        # Calculate accuracy and precision
        accuracy, precision = calculate_accuracy_precision(query_text, top_relevant_document_texts)
        
        # Store individual accuracies and precisions
        individual_accuracies.append(accuracy)
        individual_precisions.append(precision)
        
        # Update average accuracy and precision
        avg_accuracy += accuracy
        avg_precision += precision
    
    # Calculate final averages
    num_queries = len(queries_df)
    avg_accuracy /= num_queries
    avg_precision /= num_queries
    
    return avg_accuracy, avg_precision, individual_accuracies, individual_precisions

# Example usage:
avg_accuracy, avg_precision, individual_accuracies, individual_precisions = calculate_avg_accuracy_precision(queries_df, documents_df)

print("Average Accuracy:", avg_accuracy)
print("Average Precision:", avg_precision)

# You can use individual_accuracies and individual_precisions as needed.
```
